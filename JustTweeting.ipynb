# JustTweeting

We are going to build a seach engine that allows the user to find tweets based on the hashtag the user inputs. They will have an option to rank based on the topic/person being tweeted.  

```python
import csv
import re
import HTMLParser
import operator


hashTag_list = []
userName_list = []
tweet_list = []


class TweetInfo(object): 
    def __init__(self, hashtag=None, username=None, tweet=None):
        self.hashtag = hashtag
        self.username = username
        self.tweet = tweet


with open("data2/testdata.manual.2009.06.14.csv") as f:
    html_parser = HTMLParser.HTMLParser() 
    r = csv.reader(f, delimiter=',', quotechar='"')
    rgx = re.compile(r'[^"][!-@a-zA-Z0-9]+\b') 
    
    corpus = []

    for choice in r: 

        hashTag_list.append(" ".join(re.findall(rgx, choice[3]) ))
        userName_list.append(" ".join(re.findall(rgx, choice[4])))
        temp = html_parser.unescape(" ".join(re.findall(rgx, choice[5])))
        tweet_list.append(temp)



    index = 0; 
    for i in userName_list: 

        corpus.append(TweetInfo(hashTag_list[index], userName_list[index], tweet_list[index])) #add a new object to the corpus list with the needed information
        index += 1


    corpus.sort(key=  lambda mbr: operator.attrgetter('hashtag')(mbr).lower(), reverse = False) #sorts the corpus based on the hashtag 'value'


with open("orderOutput.csv", "wb") as f: 
    writer = csv.writer(f)

    for tweet in corpus: 
        writer.writerow([tweet.hashtag, tweet.username, tweet.tweet ])
```

```python
from collections import Counter

with open("orderOutput.csv", "wb") as f: 
    writer = csv.writer(f)

def get_tf(corpus):
    tf = Counter()
    for tweet in corpus:
        for word in tweet.tweet.split():
            tf[word] += 1
    return tf

tf = get_tf(corpus)
print(tf)
```

```python
with open("orderOutput.csv", "wb") as f: 
    writer = csv.writer(f)

def get_lexicon(corpus):
    lexicon = set()
    for doc in corpus:
        lexicon.update([word for word in doc.split()])
    return list(lexicon)
    
test_corpus = ['photograpy','LOVE','nike']
lexicon = get_lexicon(test_corpus)
```

```python
lexicon = get_lexicon(test_corpus)

tfm =[]
for doc in test_corpus:
    tfv = [0]*len(lexicon)
    for term in doc.split():
        tfv[lexicon.index(term)] += 1
    tfm.append(tfv)
    
print(tfm)
```

```python
import pandas as pd
from bokeh.plotting import figure, output_notebook, show, vplot

n = []
s = []
for i in range(100,1000,100):
    for tweet in corpus:
        corpus = tweet
    tfm, lexicon = get_tfm(corpus)
    c =[ [x.count(0), x.count(1)] for x in tfm]
    n_zero = sum([ y[0] for y in c])
    n_one = sum([ y[1] for y in c])  
    s.append(1.0 - (float(n_one) / (n_one + n_zero)))
    n.append(i)
    
output_notebook(hide_banner=True)
p = figure(x_axis_label='Documents', y_axis_label='Sparsity',
          plot_width=400, plot_height=400)
p.line(n, s, line_width=2)
p.circle(n, s, fill_color="white", size=8)
show(p)
```

```python
def create_inverted_index(corpus):
    idx={}
    for i, tweet in corpus:
        for word in tweet.tweet.split:
            if word in idx:
                idx[word].append(i)
            else:
                idx[word] = [i]
    return idx

test_corpus = ['love','nike','shoes']
idx = create_inverted_index(test_corpus)
```
